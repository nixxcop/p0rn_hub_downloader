```
import subprocess
import sys
import importlib



rp = [
    "yt-dlp>=2024.7.1",
    "tqdm>=4.66.1",
    "colorama>=0.4.6"
]


def install():

    for pack in rp:
        pack_name = pack.split(">=")[0]
        try:


            importlib.import_module(pack_name)
        except ImportError:

            print(f"üì¶ Package not found. Installing...")
            subprocess.check_call([sys.executable, "-m", "pip", "install", pack])

        else:
            print(f"‚úÖ Package is already installed....")


install()


import os
import yt_dlp
from tqdm import tqdm
import logging
from colorama import init, Fore


init(autoreset=True)



out = "/content/drive/MyDrive/pornhub_download"
log = "download_log.txt"

os.makedirs(out, exist_ok=True)

logging.basicConfig(
    filename=log,
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)


def phh(d):
    if d['status'] == 'downloading':
        pr = d.get('_pr_str', '').strip()
        speed = d.get('_speed_str', '')
        eta = d.get('_eta_str', '')
        tqdm.write(f"{Fore.CYAN}üì• Downloading: {pr} | Speed: {speed} | ETA: {eta}", end='\r')
    elif d['status'] == 'finished':
        tqdm.write(f"\n{Fore.GREEN}‚úÖ Finished downloading: {d['filename']}\n")


def d_one(url: str):
    try:

        ph_opt = {
            'format': 'bestvideo+bestaudio/best',
            'outtmpl': os.path.join(out, '%(title)s.%(ext)s'),
            'phhs': [phh],
            'noplaylist': True,
            'quiet': True,
        }


        with yt_dlp.YoutubeDL(ph_opt) as ydl:

            logging.info(f"Started downloading: {url}")

            ydl.download([url])

            logging.info(f"Completed: {url}")


            print(f"{Fore.GREEN}‚úÖ Download completed successfully!\n")


    except Exception as e:


        logging.error(f" Error downloading {url}: {e}")

        print(f"{Fore.RED}  Error downloading video: {e}")



if __name__ == "__main__":
    print("="*50)
    print(f"{Fore.MAGENTA}üé¨  Cornhub video downloader...!!!!!!!!! )")
    print("="*50)
    print("1Ô∏è‚É£ Download single video")
    print("2Ô∏è‚É£ Download multiple videos!!\n")
    choice = input("Enter  (1 or 2): ").strip()
    if choice == "1":
        link = input("Enter Phub video URL: ").strip()
        d_one(link)
    elif choice == "2":
        print(" working on multi downloader.....!!!")

    else:
        print(f"{Fore.RED}‚ùå Invalid choice! Exiting...")
```
